```python
import pandas as pd
```

```python
Q_YEAR = 2016  # Change me!
q1 = pd.read_excel(f"train_data/Lists ON/List ON_{Q_YEAR}/format_q1_{Q_YEAR}.xlsx")
q2 = pd.read_excel(f"train_data/Lists ON/List ON_{Q_YEAR}/format_q2_{Q_YEAR}.xlsx")
q3 = pd.read_excel(f"train_data/Lists ON/List ON_{Q_YEAR}/format_q3_{Q_YEAR}.xlsx")
q4 = pd.read_excel(f"train_data/Lists ON/List ON_{Q_YEAR}/format_q4_{Q_YEAR}.xlsx")

q2.drop(columns=["address", "build_date", "balance_accept_date", "otdel"], inplace=True)
q3.drop(columns=["address", "build_date", "balance_accept_date", "otdel"], inplace=True)
q4.drop(columns=["address", "build_date", "balance_accept_date", "otdel"], inplace=True)
```

```python
newq1q2q3q4 = (
    q1.merge(q2, on="cad_code", right_index=False)
    .merge(q3, on="cad_code", right_index=False)
    .merge(q4, on="cad_code", right_index=False)
)
```
