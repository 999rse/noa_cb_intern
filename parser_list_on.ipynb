```python
import pandas as pd  # pip install xlrd

pd.set_option("display.max_colwidth", None)
```

```python
Q_LON = 4  # Change me!
Q_YEAR = 2016  # Change me!
path2file = f"train_data/Lists ON/List ON_{Q_YEAR}/{Q_LON} quarter {Q_YEAR}.xls"
path2outfile = f"train_data/Lists ON/List ON_{Q_YEAR}/format_q{Q_LON}_{Q_YEAR}.xlsx"
```

```python
file = pd.read_excel(path2file)
```

```python
file = file.iloc[1:]
```

```python
file.tail(5)
```

```python
# Удаление строк "Итого:"
file.drop(file[file.iloc[:, 0] == "Итого:"].index, inplace=True)
```

```python
file.tail(5)
```

```python
# Удаление шапки и футера, переиндексация
file = file[8 : file.shape[0] - 2].reset_index(drop=True)
```

```python
file.shape
```

```python
otdel_names = []  # Имена отделений
index_otdel = []  # Их индексы
```

```python
# Если в строке только 1 not NaN, то она является "отделением"
for index in range(file.shape[0]):
    if (
        file.iloc[index, :].first_valid_index()
        == file.iloc[index, :].last_valid_index()
    ):
        index_otdel.append(index)
        otdel_names.append(
            file.iloc[index, :].value_counts(dropna=True).index.to_list()[0]
        )
        print(file.iloc[index, :].value_counts(dropna=True).index.to_list()[0], index)
```

```python
index_otdel = pd.Series(index_otdel).drop_duplicates()
index_otdel = index_otdel.to_list()
index_otdel
```

```python
len(index_otdel), len(otdel_names)
```

```python
def generate_list(index_otdel, otdel_names):
    result_list = []

    for i in range(len(index_otdel)):
        start_index = index_otdel[i]
        end_index = index_otdel[i + 1] if i < len(index_otdel) - 1 else len(otdel_names)
        element = otdel_names[i]

        result_list.extend([element] * (end_index - start_index))

    return result_list


result_otdel = generate_list(index_otdel, otdel_names)

# Добавляем остатки, которые не вошли в цикл generate_list
last_pack_name_numb = file.shape[0] - len(result_otdel)

result_otdel.extend([otdel_names[-1]] * last_pack_name_numb)
```

```python
len(result_otdel)
```

```python
df_otdel = pd.DataFrame(result_otdel.copy(), columns=["otdel"])
```

```python
df_otdel
```

```python
# Соединяем 2 датафрейма исходный и переформатированый под отделы
file = pd.concat([file, df_otdel], axis=1)
```

```python
file = file.iloc[:, 1:]
```

```python
# Заполняем значениями выше по строке
def fill_values_below(df):
    df = df.fillna(method="ffill")
    return df


fill_df = fill_values_below(file)
```

```python
"""
Наименование, адрес, реестровый номер объекта
Год постройки
Дата принятия на баланс Банка России
Общая площадь объекта недвижимости,  кв. м
Общая площадь здания кв. м	

Занято службами Банка России	
Не занято (не используется)	
Сдается Банком России в аренду	
Арендуется Банком России	
Сдается Банком России в субаренду	
Передано Банком России по договору безвозмездного пользования	
Передано Банку России по договору безвозмездного пользования	

Занято службами Банка России	
Не занято (не используется)	
Сдается Банком России в аренду	
Арендуется Банком России	
Сдается Банком России в субаренду	
Передано Банком России по договору безвозмездного пользования	
Передано Банку России по договору безвозмездного пользования

Балансовая стоимость, руб.	
в том числе переоценка, руб.	
Сумма начисленной амортизации, руб.	
Остаточная стоимость, руб.	
Техническое состояние
Отделение (по факту регион)
"""

# Добавляем названия колонкам
col_names = [
    "address",
    "build_date",
    "balance_accept_date",
    f"com_sq_estate_q{Q_LON}",
    f"com_sq_building_q{Q_LON}",
    f"busy_cb_1_q{Q_LON}",
    f"not_busy_cb_1_q{Q_LON}",
    f"rent_out_1_q{Q_LON}",
    f"rented_1_q{Q_LON}",
    f"rent_out_sub_1_q{Q_LON}",
    f"transfer_1_q{Q_LON}",
    f"transfer_2_q{Q_LON}",
    f"busy_cb_2_q{Q_LON}",
    f"not_busy_cb_2_q{Q_LON}",
    f"rent_out_2_q{Q_LON}",
    f"rented_2_q{Q_LON}",
    f"rent_out_sub_2_q{Q_LON}",
    f"transfer_3_q{Q_LON}",
    f"transfer_4_q{Q_LON}",
    f"balance_cost_q{Q_LON}",
    f"over_price_q{Q_LON}",
    f"accrual_amount_q{Q_LON}",
    f"residual_value_q{Q_LON}",
    f"technical_condition_q{Q_LON}",
    "otdel",
]

fill_df.columns = col_names
```

```python
# Вытаскиваем кадастровый номер
regex_pattern = r"(\d+-\d{2}-\d{2}-\d{2}-\d{2}-\d{3})"
cad_code_df = pd.DataFrame()
cad_code_df["cad_code"] = fill_df["address"].str.extract(regex_pattern)
cad_code_df.shape
```

```python
# Удаление кадастрового номера из адреса до конкатенации

fill_df.replace(regex_pattern, "", regex=True, inplace=True)
```

```python
fill_df["address"]
```

```python
final_df = pd.concat([cad_code_df, fill_df], axis=1)

final_df.dropna(inplace=True)
```
